```python
import json
import csv
import string
import re
import time
#for tokenisation
from nltk import word_tokenize
import pandas as pd

# Initialize empty list to store tweets: tweets_data
tweets_data=[]
x=[]
y=[]
k=[]
sample=[]

print("preprocessing starts")

def getdata(data_path):
    print("retriving tweets")
    tweet_path = data_path
    tweet_file = open(tweet_path, "r")
# Read in tweets and store in list: tweets_data

    for i in tweet_file:
        try:
            tweet = json.loads(i)
            tweets_data.append(tweet)
# Print the keys of the first tweet dict
            #print(tweets_data[0].keys())
            
        except:
            continue
    processdata()


def processdata():
    print("===========================")
    print("Recovering Data Teets")
    print("===========================")
    #time.sleep(1)
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    for i in range(len(tweets_data)):
        q = tweets_data[i]['text']
        o = tweets_data[i]['id_str']
        
        #print("===========================")

        #print(q)
        
        #sub is used to remove the  space
        q = RE_EMOJI.sub(r'', q)
     #   print ("diffg") 
        print("===========================")
#remove @,: # . sybols and link
  #      print(q)
       
        i = q.translate(str.maketrans('','',string.punctuation))
        
       
    #    print("===========================")
     #   print("before appending")
        print(i)    

        x.append(i)
        k.append(o)
        
#    print("checking step1")
  #  print("===========================")
#print id
   # print(k)
    #print("checking step2")
#    print("===========================")
#print msg
   # print(x)
    print("===========================")
    print("Data Tweets Recovered")
    print("===========================\n\n")
    
    
def readdict(data_path):
    print("===========================")
    print("Reading Dictionary")
    print("===========================")  
    with open(data_path) as tsvfile:
      reader = csv.reader(tsvfile, delimiter='\t')
      for row in reader:
          i = []
          i.append(row[2])
          i.append(row[5])
          y.append(i)
    print (y)        
    print("===========================")
    print("Dictionary Preparation Done")
    print("===========================\n\n")  
    addpolarity()
    
def addpolarity():  
    start_time = time.time()
    counter = 0
    print("===========================")
    print("Processing please wait...")
    print("===========================\n\n")
    
         
    for j in x:
 
            tweet_token = j
            token = word_tokenize(tweet_token)
            print("==================tokens===============================")
            print(token)
            sumnum = 0
            sum_word = 0
            for t in token:
     
                for d in y:
                    if t == d[0]:
                        sentiment = d[1]
                        if sentiment == "positive":
                            sumnum += 1
                            sum_word += 1

                        elif sentiment == "negative":
                            sumnum += -1
                            sum_word += 1

                        else:
                            sumnum += 0
                            sum_word += 1


                        break
                 
            
            if sum_word != 0.0:
                sum_more = sumnum / sum_word
                if sum_more >= 0.2:
                    sum_more = 1
   
                elif (sum_more < 0.2) and (sum_more > -0.5):
                    sum_more = 0
                   
                elif sum_more <= -0.5:
                    sum_more = -1
                   
                else:
                    print("****")
                    
                
            sum_var = []    
            varid = k[counter]
            sum_var.append(varid)
            sum_var.append(sum_more)
            sample.append(sum_var)
            counter += 1
            
            #print("===========================")

           # print(sample)
            
    print("Processing time: ", round((time.time() - start_time),8), "Seconds \n\n")
    
    time.sleep(3)
        
    print("===========================")
    print("Processing Finish")
    print("===========================")
    
    
    savetoxlsx()
    
def savetoxlsx():
    print("checking")
    df = pd.DataFrame(data=sample)
   # writer = pd.ExcelWriter('/project/processed_data/output.xlsx',engine='xlsxwriter', header=("id","sentiment"), index=False)
    
    df.to_excel('/home/user/project/processed_datacheck/output.xlsx', header=("id","sentiment"), index=False)
    
    #df.to_excel(writer, sheet_name=output.xlsx)
    #writer.save()
    
    
    #file = open("testfile_data.txt","w") 
    #file.write(some_milby) 
    #file.close() 
    
    print("===========================")
    print("Data Saved!")
    print("===========================") 
    
    
def runall():
    getdata('/home/user/project/datacheck/tweetdatacheck.txt')
    readdict('/home/user/project/datacheck/dictionarycheck.tsv')
    
runall()

    
```
