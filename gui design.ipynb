```python

from tkinter import *
import tkinter as tk
#from tkinter.tk 
#import *

root = Tk()
root.geometry("500x500")

label = Label(root, text="Hello World")
label.pack()
mytext = tk.Text(root)
mytext.insert('1.0', "I love my text widget!")
mytext.pack()
tvar = tk.StringVar()
def swaptext():
    tvar.set('There' if tvar.get() == 'Hi' else 'Hi')

my_button = tk.Button(root, textvariable=tvar, command=swaptext)
my_button.pack()

root.mainloop()
```
